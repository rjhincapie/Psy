```python
tabla=tabla.drop (["Unnamed: 0", "DateType", "InjuryCity", "InjuryCounty", "DeathCity", "DeathCounty", "Location", "LocationifOther", "OtherSignifican", "InjuryState", "ResidenceCityGeo", "InjuryCityGeo", "ResidenceState", "ResidenceCounty"], 
                  axis=1)
```

```python
tabla=tabla.drop ([0], axis=0)
```

```python
tabla.head()
```

```python
tabla.loc[tabla["Sex"] == "Male", "Sex"] = 1
tabla.loc[tabla["Sex"] == "Female", "Sex"] = 0
tabla.loc[tabla["MannerofDeath"] == "NaN", "MannerofDeath"] = "Accident"
tabla.loc[tabla["DescriptionofInjury"] == "substance abuse", "DescriptionofInjury"] = "Substance Abuse"
tabla.loc[tabla["DescriptionofInjury"] == "Substance abuse", "DescriptionofInjury"] = "Substance Abuse"
tabla.loc[tabla["Race"] == "Asian Indian", "Race"] = "Asian"
tabla.loc[tabla["Race"] == "Asian, Other", "Race"] = "Asian"
tabla.loc[tabla["Race"] == "Chinese", "Race"] = "Asian"
tabla.loc[tabla["Race"] == "Hispanic, Black", "Race"] = "Hispanic"
tabla.loc[tabla["Race"] == "Hispanic,White", "Race"] = "Hispanic"
tabla.loc[tabla["Race"] == "Native American, Other", "Race"] = "Other"
tabla.loc[tabla["Race"] == "Unknown", "Race"] = "Other"
tabla.loc[tabla["InjuryPlace"] == "Unknown", "InjuryPlace"] = "Other"
tabla.dropna(subset=["InjuryPlace", "Race"], axis=0, inplace=True)


```

```python
tabla.fillna(0, inplace=True)
tabla.drop(["COD"], axis=1, inplace=True)
```

```python
tabla['DescriptionofInjury'] = tabla['DescriptionofInjury'].replace (['0', method='pad'], inplace=True)

```
